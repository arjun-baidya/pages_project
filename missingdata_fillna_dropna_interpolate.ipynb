```python
# Import dataset
import pandas as pd
df=pd.read_csv("practiceData.csv", parse_dates=["day"])
df
```

```python
# Now I apply fillna method for different columns

new_df=df.fillna({
    'temperature':0,'windspeed':0,'event':'no event'
})
new_df
```

```python
# Now i apply forword fillna method

new_df=df.fillna(method="ffill")
new_df
```

```python
#apply backword method

new_df= df.fillna(method="bfill")
new_df
```

```python
new_df= df.fillna(method="bfill",limit=1)
new_df
```

```python
# Apply Interpolate Method

new_df= df.interpolate()
new_df
```

```python
# Dropna Method(which cells will be empty that cell will be delete)

newdf=df.dropna()
newdf
```

```python
# Dropna Method(The entire row will be delete if at least 2 column data haven't)

newdf=df.dropna(thresh=2)
newdf
```

```python
# Missing date value insert method

df=pd.date_range("2019-01-01","2019-01-11")
idx=pd.DatetimeIndex(df)
df.reindex(idx)
df
```

```python
#Handle Missing Value By Replace Function And negative value converted o Zero

import pandas as pd
import numpy as np
df=pd.read_csv("Practicedata1.csv")
df
```

```python
ndf=df.replace({-999:np.NaN,
                -88:np.NaN,
                -87:np.NaN,
                -99:np.NaN,
                -66:np.NaN,
                'No Event':'Sunny'
               
               })
ndf
```
